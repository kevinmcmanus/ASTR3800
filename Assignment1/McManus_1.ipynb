```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def MyPlot(x,y,title, my_ax=None):
    if my_ax is None:
        fig, ax = plt.subplots( figsize=(8,6))
    else:
        ax = my_ax
        
    ax.plot(x,y)
    ax.set_title(title)
    ax.set_xlabel('Variable X'); ax.set_ylabel('Variable Y')
    return ax
        
def MyErrorPlot(x,y,title, yerror, my_ax=None):
    if my_ax is None:
        fig, ax = plt.subplots( figsize=(8,6))
    else:
        ax = my_ax

    ax.errorbar(x,y, yerr=yerror)
    ax.set_title(title)
    ax.set_xlabel('Variable X'); ax.set_ylabel('Variable Y')
    return ax
    
```

```python
xx = np.linspace(0,4*np.pi, 60)
yy = np.cos(xx)
title_str='Cosine Function'

MyPlot(xx,yy, title_str)
```

```python
yerror = yy*0.2 # error is 20% of the value
ax = MyErrorPlot(xx,yy,  title_str+' (with Error Bars)', yerror=yerror)
ax.grid(axis='y')
```
